# Perceptron

```python
import numpy as np
class Perceptron(object):
    def __init__(self, X, Y,epoches,learning_rate):
        self.X = X #traning samples
        self.Y = Y  #results
        self.W = [] #weight
        self.b = None #bias
        self.learning_rate = learning_rate #
        self.cnt = 0
        self.epoches = epoches
        self.train()
    
    def train(self):
        self.cnt += 1
        #init w and b
        self.W = np.full(shape=np.shape(self.X)[1], fill_value=0.0)
        self.b = 0
        
        #for i in range(np.shape(self.X)[0]):
           # if (self.cnt>self.epoches):
            # break
            #else:
        for i in range(self.epoches):
             for i in range(np.shape(self.X)[0]):
                loss = self.Y[i] * (np.dot(self.W, self.X[i]) + self.b)
                if loss <= 0: #classification correct
                    self.W = self.W + self.learning_rate*self.X[i]*self.Y[i]
                    self.b = self.b + self.learning_rate*self.Y[i]
        
        return self.W,self.b,self.cnt

    def Classify(self, test_samples):
        sign = lambda x : (np.sign(x)<0 and [-1] or [+1])[0]
        result = [sign(np.dot(self.W, x) + self.b) for x in X]
        return result
```

# Use different epoches in perceptron

```python
from sklearn.metrics import accuracy_score
y=np.array(y)
X=np.array(X)
scorelist=[]
xlist=[]
for i in range(100):
    p = Perceptron(X,y,i,1)
    res = p.Classify(X)
    score = accuracy_score(y,res)
    xlist.append(i)
    scorelist.append(score)
```

```python
print(scorelist)
```

```python
import matplotlib.pyplot as plt
#plot the score curve
plt.plot(xlist,scorelist)
plt.axis
plt.xlabel('Score',fontproperties='SimHei',fontsize=20,color='g')
plt.ylabel('Epoches',fontproperties='SimHei',fontsize=20)
plt.title(r'Score Curve',fontproperties='SimHei',fontsize=20)
plt.grid(True)
```
